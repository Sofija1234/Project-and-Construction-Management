```python
# Importing necessary libraries
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from google.colab import files
import io

# Uploading the CSV file
uploaded = files.upload()

# Reading the uploaded file
file_name = list(uploaded.keys())[0]

# Reading the CSV file into a Pandas DataFrame
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# Checking the first few rows to confirm the file is loaded correctly
print(df.head())

# Calculating Spearman Rank Correlation Matrix
spearman_corr = df.corr(method='spearman')

# Displaying the Spearman Rank Correlation Matrix
print(spearman_corr)

# Visualizing the Spearman correlation matrix using a heatmap
import seaborn as sns
plt.figure(figsize=(14, 10))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Spearman Rank Correlation Matrix')
plt.show()

```
